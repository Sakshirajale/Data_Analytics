# Merging. Joining and Concatenating
There are 3 main ways of combining DataFrame together:Merging, joining and Concatinating. In this lecture we will discuss these
3 methods with examples.

# Examples DataFrame

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'], 
                    'B': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C4', 'C5', 'C6', 'C7'], 
                    'D': ['D4', 'D5', 'D6', 'D7']}, index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'], 
                    'B': ['B8', 'B9', 'B10', 'B11'], 
                    'C': ['C8', 'C9', 'C10', 'C11'], 
                    'D': ['D8', 'D9', 'D10', 'D11']}, index=[8, 9, 10, 11])
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 
                     'A': ['A0', 'A1', 'A2', 'A3'], 
                     'B': ['B0', 'B1', 'B2', 'B3']}) 
```

```python
left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 
                      'C': ['C0', 'C1', 'C2', 'C3'], 
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
right
```

```python
pd.merge(left,right,how="inner",on="key")
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'], 'key2': ['K0', 'K1', 'K0', 'K1'], 'A': ['A0', 'A1', 'A2', 'A3'], 'B': ['B0', 'B1', 'B2', 'B3']}) 
```

```python
left
```

```python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'], 'key2': ['K0', 'K0', 'K0', 'K0'], 'C': ['C0', 'C1', 'C2', 'C3'], 'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
right
```

```python
pd.merge(left,right,how="inner",on=["key1","key2"])
```

```python
pd.merge(left,right,how="left",on=["key1","key2"])
```

```python
pd.merge(left,right,how="right",on=["key1","key2"])
```

```python
pd.merge(left,right,how="outer",on=["key1","key2"])
```

```python
pd.merge(left,right,how="cross") #each row frame match with right datqframe
```

# Joining
It is a convient methods for combining the columns 

```python
data1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']}, index=['K0', 'K1', 'K2']) 
```

```python
data2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'], 'D': ['D0', 'D2', 'D3']}, index=['K0', 'K2', 'K3'])
```

```python
data1
```

```python
data2
```

```python
data1.join(data2) #join two table
```

```python
data2.join(data1)
```

```python
data1.join(data2,how="right")
```

```python
data1.join(data2,how="outer")
```

```python
data1.join(data2,how="inner")
```
